```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data with a custom line style and marker
ax.plot(x, y, color='#5D1725', linewidth=2.5, linestyle='-', marker='o', markersize=6, markerfacecolor='orange', label='Sine Wave')

title_font = {'family': 'DejaVu Sans', 'color': '#333333', 'weight': 'bold', 'size': 22}
label_font = {'family': 'DejaVu Sans', 'color': '#333333', 'weight': 'medium', 'size': 16}


# Set the title and labels with custom fonts
ax.set_title('Pretty Sine Wave Plot', fontsize=20, fontdict=title_font, pad = 20)
ax.set_xlabel('X Axis (Time)', fontsize=16, fontweight='medium', color='#333333')
ax.set_ylabel('Y Axis (Amplitude)', fontsize=16, fontweight='medium', color='#333333')

# Customize the grid
ax.grid(True, which='both', linestyle='--', linewidth=0.6, color='gray', alpha=0.7)

# Customize the ticks
ax.tick_params(axis='both', which='major', labelsize=12, width=1.5, length=7, color='#333333')
ax.tick_params(axis='both', which='minor', labelsize=10, width=1, length=4, color='#333333')

# Add a legend with a custom location and styling
ax.legend(loc='lower right', fontsize=14, frameon=True, shadow=True, borderpad=1)

# Customize the spines (the borders of the plot)
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1.5)
    spine.set_color('#333333')

# Add a tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
