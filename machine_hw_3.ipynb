# Problem 5-1a

```python
Sy = 37.5 #kpsi

s_1 = 25
s_2 = 15
tau_xy = 0

def vonMises(sig_x1, sig_y1, tau):
    return ((sig_x1**2 - sig_x1*sig_y1 + sig_y1**2 + 3*tau**2)**(1/2))



ans_vm = vonMises(s_1, s_2, tau_xy)
print(ans_vm, " ksi")

# Getting Factor of safety for von mises is easy since we do not need to find any prinicpal stresses

n_vm = Sy/ans_vm # Von Mises
n_MSS = Sy/s_1
print("The VM Factor of Safety is: ", n_vm)
print("The MSS Factor of Safety is: ", n_MSS)

```
